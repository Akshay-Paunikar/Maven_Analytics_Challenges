#### Trend of Summits & Deaths Over Time

```python
# Group data by year
yearly_stats = everest.groupby('year').agg({
    'msuccess': 'sum',
    'death': 'sum'
}).reset_index()

# Group data by season
seasonal_stats = everest.groupby('season').agg({
    'msuccess': 'sum',
    'death': 'sum'
}).reset_index()
```

```python
seasonal_stats
```

```python
# show me the top 5 year with maximum summits
yearly_stats.sort_values(by=['msuccess'], ascending=False).head(100).reset_index(drop=True)
```

```python
yearly_stats.sort_values(by=['death'], ascending=False).head(100).reset_index(drop=True)
```

```python
# Plot 1: Summits and Deaths Over Time
plt.figure(figsize=(15, 7))
plt.subplot(2, 1, 1)
sns.lineplot(data=yearly_stats, x='year', y='msuccess', label='Summits', marker='o')
sns.lineplot(data=yearly_stats, x='year', y='death', label='Deaths', marker='o', color='red')
plt.title('Summits and Deaths on Everest Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.legend()

# Plot 2: Seasonal Distribution of Summits and Deaths
plt.subplot(2, 1, 2)
season_melted = pd.melt(seasonal_stats, id_vars='season', value_vars=['msuccess', 'death'],
                        var_name='Outcome', value_name='Count')
sns.barplot(data=season_melted, x='season', y='Count', hue='Outcome', palette='Set2')
plt.title('Seasonal Distribution of Summits and Deaths')
plt.xlabel('Season')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

From the line chart showing summits and deaths per year:

Summits have increased significantly since the 1990s, peaking in:

 -  2019: 872 summits

 - 2024: 839 summits

 - 2018: 810 summits

Death peaks are relatively low compared to summits:

 - 2023 saw the highest death count in recent years (18 deaths).

 - 2015 is notable for having 0 summits but 14 deaths, likely due to the Nepal earthquake disaster.

Spring dominates Everest climbs, offering the most favorable weather windows and thus highest summit and death counts.

#### B: Success Rates by Nationality & Gender

```python
# Group by nationality and gender
nationality_gender_stats = everest.groupby(['citizen', 'sex']).agg(
    attempts=('msuccess', 'count'),
    successes=('msuccess', 'sum'),
    deaths=('death', 'sum')
).reset_index()

# Calculate rates
nationality_gender_stats['success_rate'] = (nationality_gender_stats['successes'] / nationality_gender_stats['attempts']) * 100
nationality_gender_stats['death_rate'] = (nationality_gender_stats['deaths'] / nationality_gender_stats['attempts']) * 100

# Focus on top 15 countries by number of attempts
top_countries = nationality_gender_stats.groupby('citizen')['attempts'].sum().sort_values(ascending=False).head(15).index
top_nationality_gender_stats = nationality_gender_stats[nationality_gender_stats['citizen'].isin(top_countries)]
```

```python
# Plot success rates
plt.figure(figsize=(16, 5))
sns.barplot(data=top_nationality_gender_stats, x='citizen', y='success_rate', hue='sex')
plt.title('Success Rate by Nationality & Gender')
plt.xlabel('Nationality')
plt.ylabel('Success Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot death rates
plt.figure(figsize=(16, 5))
sns.barplot(data=top_nationality_gender_stats, x='citizen', y='death_rate', hue='sex', palette='Set2')
plt.title('Death Rate by Nationality & Gender')
plt.xlabel('Nationality')
plt.ylabel('Death Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
